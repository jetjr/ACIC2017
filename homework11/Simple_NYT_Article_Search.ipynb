# NYT API ARTICLE SEARCH

## Instructions: 

Edit the search variable to include a list of keyword terms to search for New York Times Articles. 

Examples: search = ['mars', 'earth', 'mercury'] 
          search = ['tucson', 'phoenix', 'flagstaff']

```python
import requests
import json

search = ['mars', 'earth', "mercury"]

for i in search:
    target = 'http://api.nytimes.com/svc/search/v2/articlesearch.json?q={}&api-key=a3597c6a331f490abe9b72f1b9614979'.format(i)
    r = requests.get(url = target)
    x = json.loads(r.text)
    print(json.dumps(x, indent = 4))
```
